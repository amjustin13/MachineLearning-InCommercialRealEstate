```python
import pandas as pd
import numpy as np
import os
dataset = pd.read_csv('datasets/Issued_Construction_Permits_Austin.csv', low_memory = False, encoding='utf-8')
```

```python
dataset.head()
```

```python
df = pd.DataFrame(dataset)
cols = [1, 8]
df = df[df.columns[cols]]
df = df.rename(columns={'Permit_Type_Desc': 'permit_type_description', 'Description': 'description'})
df.head()
```

```python
data_size = df.shape[0]
print(f"Number of records before removing duplicates: {data_size}")
df.drop_duplicates(subset=['description'])
data_size_after_drop = df.shape[0]
removed = data_size - data_size_after_drop
print(f"Removed {removed} duplicates.")
```

```python
df = df.replace(' ', np.nan)
nan_values = df[df.isna().any(axis=1)]
nan_values
```

```python
df.dropna()
```

```python
df['permit_type_description'].value_counts()
```

```python
os.makedirs("austin_ds")
```

```python
label1 = "Electrical Permit"
label2 = "Plumbing Permit"
label3 = "Building Permit"
label4 = "Mechanical Permit"
label5 = "Driveway / Sidewalks"
labels = [label1, label2, label3, label4,label5]
directories = ["electrical_permit", "plumbing_permit", "building_permit", "mechanical_permit", "driveway_sidewalk_permit"]

for directory in directories:
    os.makedirs("austin_ds/" + directory)   
```

```python
count1, count2, count3, count4, count5 = 0, 0, 0, 0, 0
for i in range(len(df)):
    if (df.iloc[i].permit_type_description == label1):
        file = open("austin_ds/electrical_permit/" + str(count1) + ".txt", 'w', encoding="utf-8")
        file.write(str(df.iloc[i].description))
        file.close()
        count1 += 1
    if (df.iloc[i].permit_type_description == label2):
        file = open("austin_ds/plumbing_permit/" + str(count2) + ".txt", 'w', encoding="utf-8")
        file.write(str(df.iloc[i].description))
        file.close()
        count2 += 1
    if (df.iloc[i].permit_type_description == label3):
        file = open("austin_ds/building_permit/" + str(count3) + ".txt", 'w', encoding="utf-8")
        file.write(str(df.iloc[i].description))
        file.close()
        count3 += 1
    if (df.iloc[i].permit_type_description == label4):
        file = open("austin_ds/mechanical_permit/" + str(count4) + ".txt", 'w', encoding="utf-8")
        file.write(str(df.iloc[i].description))
        file.close()
        count4 += 1
    if (df.iloc[i].permit_type_description == label5):
        file = open("austin_ds/driveway_sidewalk_permit/" + str(count5) + ".txt", 'w', encoding="utf-8")
        file.write(str(df.iloc[i].description))
        file.close()
        count5 += 1
```
